```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('heart_disease_cleaned.csv')
df
```

```python
# Remove unnamed column

if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
```

```python
# Target Variable Distribution

plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=df)
plt.title('Heart Disease Presence (Target)')
plt.xlabel('Target (0: No Disease, 1: Disease)')
plt.ylabel('Count')
plt.show()
```

Interpretation:
Class 0 has more patients than class 1, but the distribution is relatively balanced.


```python
# Age Distribution

plt.figure(figsize=(6, 4))
sns.histplot(df['age'], kde=True, bins=15)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Interpretation:
Most patients are aged 45â€“60. Age is slightly right-skewed.

```python
# Cholesterol Distribution

plt.figure(figsize=(6, 4))
sns.histplot(df['chol'], kde=True, bins=20)
plt.title('Cholesterol Level Distribution')
plt.xlabel('Cholesterol (mg/dL)')
plt.ylabel('Frequency')
plt.show()
```

Interpretation:
Most patients have a cholesterol level around 350 mg/dL, as shown by the highest bar in the histogram.
Cholesterol levels are unevenly distributed across patients, with values ranging widely from 120 to 550 mg/dL.

```python
# Resting Blood Pressure (trestbps)

plt.figure(figsize=(6, 4))
sns.boxplot(x='target', y='trestbps', data=df)
plt.title('Resting Blood Pressure vs Heart Disease')
plt.show()
```

Interpretation:
Patients with heart disease (target=1) show slightly higher blood pressure, but overlap is significant.


```python
# Maximum Heart Rate (thalach)

plt.figure(figsize=(6, 4))
sns.boxplot(x='target', y='thalach', data=df)
plt.title('Max Heart Rate vs Heart Disease')
plt.show()
```

Interpretation:

Higher heart rates (thalach) are observed more in heart disease cases.


```python
# Count Plot for Sex

plt.figure(figsize=(6, 4))
sns.countplot(x='sex', data=df, hue='target')
plt.title('Heart Disease Count by Sex (0: Female, 1: Male)')
plt.show()
```

Interpretation:

Females (sex = 0) have a higher overall count, and more females do not have heart disease (target = 0).

Males (sex = 1) show nearly equal counts of those with and without heart disease.

```python
# Count Plot for Chest Pain Type (cp)

plt.figure(figsize=(6, 4))
sns.countplot(x='cp', data=df, hue='target')
plt.title('Chest Pain Type vs Heart Disease')
plt.show()
```

Interpretation:
Type 2 chest pain (asymptomatic) is most strongly associated with heart disease.

```python
# Correlation Heatmap

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```

Interpretation:

'cp', 'thalach', and 'slope' have positive correlation with heart disease.
'exang', 'oldpeak', and 'ca' are negatively correlated.
